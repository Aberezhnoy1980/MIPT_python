### Задача 1. 
Оценить вероятносять того, что на игральной кости выпадет больше 4-х очков. Сравнить с теоретической вероятностью

```python
trials = 100000 # количесттво попыток
success = 0 # количество успехов

for i in range(trials):
    random_score = np.random.randint(1, 7)
    if random_score > 4:
        success += 1
        
p_monte_carlo = success / trials # вероятность из модделирования
p = 1 / 3                        # классическая вероятность 

print(success, trials, p_monte_carlo, p)
```

### Задача 1А. 
Дан квадрат, в который вписан круг (касаясь сторон квадрата). В квадрат случайным образом попадают точки. Какова вероятность, что наугад взятая точка попадёт в круг?
Подсказка: Удобно взять квардат со стороной, равной 2, и отсчитывать координаты точек от его центра.

```python
trials = 1000000
successes = 0

for i in range(100, 1000):
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(-1, 1)

    if x**2 + y**2 < 1:
        successes += 1

res = successes / trials
print (res)
```
